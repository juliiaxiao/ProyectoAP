```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 1. Carga de datos
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Para el EDA usamos el set de entrenamiento
X = X_train
y = y_train

# Diccionario de etiquetas para visualización
label_names = ['Camiseta', 'Pantalón', 'Jersey', 'Vestido', 'Abrigo',
               'Sandalia', 'Camisa', 'Zapatilla', 'Bolso', 'Botín']

# 2. EDA: Cantidad de datos y dimensiones
print(f"Cantidad de imágenes: {X.shape[0]}")
print(f"Dimensiones de cada imagen: {X.shape[1]}x{X.shape[2]}")

# 3. Características estadísticas
print(f"Valor máximo de píxel: {X.max()}")
print(f"Valor mínimo de píxel: {X.min()}")
print(f"Valor medio de píxeles: {X.mean():.2f}")

# 4. Visualización (Obligatorio para el EDA)
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X[i], cmap=plt.cm.binary)
    plt.xlabel(label_names[y[i]])
plt.show()
```
